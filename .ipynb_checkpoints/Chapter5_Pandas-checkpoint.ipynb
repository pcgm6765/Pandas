# Pandas
## DataFrame
DataFrame為pandas套件內之核心功能，專門用於處理表格資料。其以column為類別與index為索引分類資料，以便後續之資料處理。

```python
import pandas as pd
df=pd.DataFrame([10, 20, 30, 40], columns=['number'], index=['a', 'b', 'c', 'd'])
df
```

可透過index或columns方法查看其索引與欄位之命名

```python
df.index
```

```python
df.columns
```

亦可選定特定索引名稱顯示其值

```python
df.loc['c']
```

```python
df.loc[['a', 'd']]
```

也可選定索引之計數以顯示對應之值（包含起始不包含最後）

```python
df.iloc[1:3]
```

可對每個column進行相加

```python
df.sum()
```

也可使用匿名函數進行操作

```python
df.apply(lambda x: x**2)
```

或直接對DataFrame進行運算操作

```python
df**2
```

增加column可使用tuple

```python
df['floats']=(1.5, 2.5, 3.5, 4.5)
df
```

也可使用DataFrame

```python
df['names']=pd.DataFrame(['Yves', 'Sandra', 'Lilli', 'Henry'], index=['d', 'a', 'b', 'c'])
df
```

可使用append搭配字典與DataFrame加入一列(pandas2.0已刪除append)

```python
df=df.append(pd.DataFrame({'numbers': 100, 'floats': 5.75, 'names': 'Jill'}, index=['y',]))
```

可對個別行取平均

```python
df[['number', 'floats']].mean()
```

可對個別行取標準差

```python
df[['number', 'floats']].std()
```

亦可從numpy ndarray產生DataFrame

```python
import numpy as np
np.random.seed(100)
a=np.random.standard_normal((9, 4))
a
```

```python
df=pd.DataFrame(a)
df
```

可設定columns name

```python
df.columns=['No1', 'No2', 'No3', 'No4']
df
```

若要處裡時間續列問題 可使用pandas內建函數自動生成時間序列之index

```python
dates=pd.date_range('2026-1-1', periods=9, freq='M')
dates
```

```python
df.index=dates
df
```

將DataFrame轉為numpy ndarray有下面兩種方法 其一為取values

```python
df.values
```

其二為使用numpy

```python
np.array(df)
```

## Group by Operation
其效果如Excel之樞紐分析 可根據不同的類別進行分類與比較

```python
df['Quarter']=['Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q3', 'Q3', 'Q3']
df
```

用Quarter進行分類 即可進行各種分析

```python
groups=df.groupby('Quarter')
groups
```

```python
groups.size()
```

```python
groups.mean()
```

```python
groups.max()
```

```python
groups.aggregate([min, max]).round(2)
```

```python
df['Odd_Even']=['Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even', 'Odd']
df
```

```python
groups=df.groupby(['Quarter', 'Odd_Even'])
groups.size()
```

```python
groups[['No1', 'No4']].aggregate([sum, np.mean])
```

## Complex Selection
可設計特殊條件對DataFrmae進行篩選

```python
data=np.random.standard_normal((10,2))
df=pd.DataFrame(data, columns=['x', 'y'])
df
```

可指定輸出前五行或後五行

```python
df.head()
```

```python
df.tail()
```

```python
(df['x']>0)&(df['y']<0)
```

```python
df[(df['x']>0)&(df['y']<0)]
```

用query可達成相同效果

```python
df.query('x>0 & y<0')
```
